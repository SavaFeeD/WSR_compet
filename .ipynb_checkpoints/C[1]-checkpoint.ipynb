```python
from sklearn.feature_extraction.text import TfidfVectorizer

# competencies_text = pd.read_json('competencies_descriptions_text.json')
competencies_text = pd.read_json('competencies_text.json')
sections_competencies_text = pd.read_json('sections_competencies_text.json')

cvector = TfidfVectorizer()
cvector_text = cvector.fit_transform(competencies_text['l'])
cvector_sections = cvector.fit_transform(sections_competencies_text['l'])
```

```python
from sklearn.cluster import KMeans
```

```python
from collections import Counter

km = KMeans(n_clusters=7, random_state=43)
idx = km.fit(cvector_text)
clusters = km.labels_.tolist()
clusters_info = Counter(clusters)
print(clusters)
print(clusters_info)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import IncrementalPCA
```

```python
dist = cosine_similarity(cvector)

icpa = IncrementalPCA(n_components=2, batch_size=70)
icpa.fit(dist)
demo2 = icpa.transform(dist)
xs, ys = demo2[:, 0], demo2[:, 1]

data = {'x': xs, 'y' : ys, 'titles': df['0']} #2D
dff = pd.DataFrame(data)
dff['cluster'] = clusters

# dff.replace(9, 'black', inplace=True)
# dff.replace(8, 'gray', inplace=True)
dff.replace(7, 'pink', inplace=True)


dff.replace(6, 'red', inplace=True)
dff.replace(5, 'blue', inplace=True)
dff.replace(4, 'green', inplace=True)
dff.replace(3, 'orange', inplace=True)
dff.replace(2, 'yellow', inplace=True)
dff.replace(1, 'magenta', inplace=True)
dff.replace(0, 'purple', inplace=True)


# print(len(xs))
fig = plt.figure(figsize=(31, 20))
# ax = fig.add_subplot(111, projection='3d')
ax = fig.add_subplot(111) # 2D
for i in range(0, dff.shape[0]):
    ax.scatter(dff['x'].iloc[i], dff['y'].iloc[i], c=dff['cluster'].iloc[i], cmap='jet', s=300)
dff.apply(lambda x: ax.annotate(x['titles'], (x['x'], x['y'])), axis=1) # Подписи для точек (название компетенции) только для 2Д
ax.set_xlabel('X')
ax.set_ylabel('Y')
# ax.set_zlabel('Z') # Для ЗД
plt.title('WSR Competencies clusters')
```

```python
ddd = pd.DataFrame(cvector_sections.toarray())
ddd['cluster'] = dff['cluster']
```

```python
ddd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split

X = ddd.drop('cluster', axis=1)
y = ddd['cluster']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
model = RandomForestClassifier(n_estimators = 100)
model.fit(X_train, y_train)
```

```python
from sklearn import metrics
```

```python
y_pred = model.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```
