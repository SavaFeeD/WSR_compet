# Предобработка датасета

<b>Меняем названия колонок у хедера на более адекватные 

```python
df = df.rename(columns={f'Unnamed: {x}': f'{x}' for x in range(0, df.shape[1])})
df.head()
```

<b>Формирую новый датафрейм с тремя колонками Название Описание и Секции на основе данных из датасета

```python
datas = []

def is_digit(elem):
    try:
        float(elem)
        return True
    except ValueError:
        return False
    
    
for i in range(df.shape[0]):
    row = pd.Series(df.iloc[i]).dropna()
    
    new_data = {
        "name": row[0].lower().strip(),
        "desc": row[1].lower().strip(),
        "sections": []
    }
    
    
    for item_i in range(row.shape[0]):
        step = 3
        if item_i%step == 0:
            if not is_digit(row[item_i]):
                step = 2
        
        if step == 3:
            if item_i%step == 0 and item_i != 0 and row[item_i] != 'Всего' and row[item_i] != 'всего' and row[item_i] != 100.0:
                schema = {
                    "title": row[item_i-1],
                    "rate": row[item_i],
                    "skills": row[item_i+1]
                }
                new_data['sections'].append(schema)
                
    datas.append(new_data)
    
new_df = pd.DataFrame(datas)
new_df
```

### Очистим текстовые данные

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
```

```python
def clean_word_tok(text):
    return word_tokenize(text)

def clean_sect_word_tok(sec):
    section = sec
    
    for key, item in enumerate(section):
        section[key]['title'] = word_tokenize(item['title'])
        section[key]['skills'] = word_tokenize(item['skills'])
    return section
```

```python
new_df['sections'] = new_df['sections'].apply(clean_sect_word_tok)
```

```python
new_df['desc'] = new_df['desc'].apply(clean_word_tok)
```

```python
new_df
```

### Собираем все слова предложения в тексты в колонке sections
#### Там же множим умения в зависимости от важности раздела

```python
def unif_sections(sections):
    all_extends = []
    
    # собираю умения и названия разделов в поле title   
    for key, section in enumerate(sections):
        section['title'].extend(section['skills'])
        sections[key]['title'] = section['title']
        
    # множу текст каждого раздела и сохраняю в массив all_extend         
    for section in sections:        
        for i in range(round(float(section['rate']))):
            all_extends.extend(section['title'])
        
    # возвращаю размноженный текст    
    return all_extends
```

```python
new_df['sections'] = new_df['sections'].apply(unif_sections)
```

### Очистим тексты от пунктуации

```python
new_df['desc'] = new_df['desc'].apply(lambda x: [word for word in x if word.isalpha()])
new_df['sections'] = new_df['sections'].apply(lambda x: [word for word in x if word.isalpha()])
```

```python
new_df
```

### Отфильтруем стоп слова

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('russian'))
```

#### Приведем все слова в нижний регистр

```python
new_df['desc'] = new_df['desc'].apply(lambda x: [word.lower() for word in x])
new_df['sections'] = new_df['sections'].apply(lambda x: [word.lower() for word in x])
```

```python
new_df
```

##### Берем тексты и оставляем все кроме стоп слов

```python
new_df['desc'] = new_df['desc'].apply(lambda x: [word for word in x if not word in stop_words])
new_df['sections'] = new_df['sections'].apply(lambda x: [word for word in x if not word in stop_words])
```

```python
new_df
```

### Сокращаем все слова до их основ

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
snowball = SnowballStemmer('russian')
```

```python
new_df['desc'] = new_df['desc'].apply(lambda x: [snowball.stem(word) for word in x])
```

```python
new_df
```

### Создаем колонку в которой мы соединим Описание и Разделы для кластеризации

```python
new_df['desc-sections'] = [[] for i in range(new_df.shape[0])]
```

```python
for i in range(new_df.shape[0]):
    new_df.iloc[i]['desc-sections'].extend(new_df.iloc[i]['desc'])
    new_df.iloc[i]['desc-sections'].extend(new_df.iloc[i]['sections'])    
```

```python
new_df.head()
```

```python
new_df['desc'] = new_df['desc'].apply(lambda x: ' '.join(x))
new_df['sections'] = new_df['sections'].apply(lambda x: ' '.join(x))
new_df['desc-sections'] = new_df['desc-sections'].apply(lambda x: ' '.join(x))
```

# Кластеризация

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.cluster import KMeans
```

```python
from collections import Counter

km = KMeans(n_clusters=7, max_iter=500)
idx = km.fit(tfidf_desc_compet)
clusters = km.labels_.tolist()
clusters_info = Counter(clusters)
print(clusters_info)
```

<b>Визуализируем кластеры

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import IncrementalPCA
```

```python
dist = cosine_similarity(tfidf_desc_compet)

icpa = IncrementalPCA(n_components=2, batch_size=70)
icpa.fit(dist)
demo2 = icpa.transform(dist)
xs, ys = demo2[:, 0], demo2[:, 1]

data = {'x': xs, 'y' : ys, 'titles': df['0']}

# Создаем отдельную колонку 'cluster' чтобы задать каждому вектору его кластер
df_with_clusters = pd.DataFrame(data)
df_with_clusters['cluster'] = clusters

# Для наглядности на графике используем замену номера кластера на определенный цвет
df_with_clusters.replace(6, 'red', inplace=True)
df_with_clusters.replace(5, 'blue', inplace=True)
df_with_clusters.replace(4, 'green', inplace=True)
df_with_clusters.replace(3, 'orange', inplace=True)
df_with_clusters.replace(2, 'yellow', inplace=True)
df_with_clusters.replace(1, 'magenta', inplace=True)
df_with_clusters.replace(0, 'purple', inplace=True)


# Визуализируем
fig = plt.figure(figsize=(31, 20))
ax = fig.add_subplot(111)

for i in range(0, df_with_clusters.shape[0]):
    ax.scatter(df_with_clusters['x'].iloc[i],
               df_with_clusters['y'].iloc[i],
               c=df_with_clusters['cluster'].iloc[i],
               cmap='jet',
               s=300)
    
df_with_clusters.apply(lambda x: ax.annotate(x['titles'], (x['x'], x['y'])), axis=1)

ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.title('WSR Competencies clusters')
```
