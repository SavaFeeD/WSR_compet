```python
competencies_text = pd.read_json('competencies_text.json')
competencies_text['l']
cvector = TfidfVectorizer()
cvector = cvector.fit_transform(competencies_text['l'])
cvector
```

```python
from sklearn.cluster import KMeans
```

```python
from collections import Counter

km = KMeans(n_clusters=7)
idx = km.fit(cvector)
clusters = km.labels_.tolist()
clusters_info = Counter(clusters)
print(clusters)
print(clusters_info)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import IncrementalPCA
```

```python
dist = cosine_similarity(cvector)

icpa = IncrementalPCA(n_components=2, batch_size=70)
icpa.fit(dist)
demo2 = icpa.transform(dist)
xs, ys = demo2[:, 0], demo2[:, 1]

data = {'x': xs, 'y' : ys, 'titles': df['0']} #2D
df = pd.DataFrame(data)

# print(len(xs))
fig = plt.figure(figsize=(31, 20))
# ax = fig.add_subplot(111, projection='3d')
ax = fig.add_subplot(111) # 2D
ax.scatter(df['x'], df['y'], s=300)
df.apply(lambda x: ax.annotate(x['titles'], (x['x'], x['y'])), axis=1) # Подписи для точек (название компетенции) только для 2Д
ax.set_xlabel('X')
ax.set_ylabel('Y')
# ax.set_zlabel('Z') # Для ЗД
plt.title('WSR Competencies clusters')
```
