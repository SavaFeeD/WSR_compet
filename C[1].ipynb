# Кластеризация

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# competencies_text = pd.read_json('competencies_descriptions_text.json')
competencies_text = pd.read_json('competencies_text.json')
sections_competencies_text = pd.read_json('sections_competencies_text.json')
all_titles_tokens = pd.read_json('all_titles_tokens.json')
all_titles_tokens
```

```python
cvector = TfidfVectorizer()
cvector_clustering = cvector.fit_transform(competencies_text['l'])
cvector_classification = cvector.fit_transform(sections_competencies_text['l'])
cvector_all_titles_tokens = cvector.fit_transform(all_titles_tokens[0])
```

```python
from sklearn.cluster import KMeans
```

```python
from collections import Counter

km = KMeans(n_clusters=7, random_state=43)
idx = km.fit(cvector_clustering)
clusters = km.labels_.tolist()
clusters_info = Counter(clusters)
print(clusters)
print(clusters_info)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import IncrementalPCA
```

<b>Визуализируем кластеры

```python
dist = cosine_similarity(cvector_clustering)

icpa = IncrementalPCA(n_components=2, batch_size=70)
icpa.fit(dist)
demo2 = icpa.transform(dist)
xs, ys = demo2[:, 0], demo2[:, 1]

# Создаем отдельную колонку 'cluster' чтобы задать каждому вектору его кластер
data = {'x': xs, 'y' : ys, 'titles': df['0']}
df_with_clusters = pd.DataFrame(data)
df_with_clusters['cluster'] = clusters

# Для наглядности на графике используем замену номера кластера на определенный цвет
df_with_clusters.replace(6, 'red', inplace=True)
df_with_clusters.replace(5, 'blue', inplace=True)
df_with_clusters.replace(4, 'green', inplace=True)
df_with_clusters.replace(3, 'orange', inplace=True)
df_with_clusters.replace(2, 'yellow', inplace=True)
df_with_clusters.replace(1, 'magenta', inplace=True)
df_with_clusters.replace(0, 'purple', inplace=True)


# Визуализируем
fig = plt.figure(figsize=(31, 20))
ax = fig.add_subplot(111)

for i in range(0, df_with_clusters.shape[0]):
    ax.scatter(df_with_clusters['x'].iloc[i],
               df_with_clusters['y'].iloc[i],
               c=df_with_clusters['cluster'].iloc[i],
               cmap='jet',
               s=300)
    
df_with_clusters.apply(lambda x: ax.annotate(x['titles'], (x['x'], x['y'])), axis=1)

ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.title('WSR Competencies clusters')
```

# Классификация

```python
df_classification = pd.DataFrame(cvector_all_titles_tokens.toarray())
df_classification['cluster'] = df_with_clusters['cluster']
```

```python
df_classification
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split

X = df_classification.drop('cluster', axis=1)
y = df_classification['cluster']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
model = RandomForestClassifier(n_estimators = 100)
model.fit(X_train, y_train)
```

```python
from sklearn import metrics
```

```python
y_pred = model.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```
